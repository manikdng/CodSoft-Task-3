```python
species_count = df['species'].value_counts()
plt.figure(figsize=(8, 5))
species_count.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
species_count = df['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140, colors=['blue', 'Pink', 'green'])
plt.title('Distribution of Iris Species')
plt.axis('equal') 
plt.show()
```

```python
sb.pairplot(df)
plt.suptitle('Pair Port of Iris Flower', y=1.00)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sb.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", style="species", palette="deep")
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid()
plt.show()
```
